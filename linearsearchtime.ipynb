```python
import time
import random
import matplotlib.pyplot as plt

def linear_search(arr, x):
    """
    Linear search algorithm to find the index of element x in arr.
    Returns -1 if element not found.
    """
    for i in range(len(arr)):
        if arr[i] == x:
            return i
    return -1

def time_linear_search(arr, x):
    """
    Measure the time taken for linear search on array arr for element x.
    """
    start_time = time.time()
    linear_search(arr, x)
    end_time = time.time()
    return end_time - start_time

def plot_search_time_vs_n(max_n, step):
    """
    Plot the graph of search time versus n.
    """
    n_values = []
    search_times = []
    for n in range(1, max_n + 1, step):
        n_values.append(n)
        arr = [random.randint(1, 1000) for _ in range(n)]
        search_time = time_linear_search(arr, random.randint(1, 1000))  # Measure search time
        search_times.append(search_time)

    plt.plot(n_values, search_times, marker='o')
    plt.title('Linear Search Time Complexity')
    plt.xlabel('Number of Elements (n)')
    plt.ylabel('Time taken for search (seconds)')
    plt.grid(True)
    plt.show()

# Example usage
max_n = 1000  # Maximum value of n
step = 50     # Step size for increasing n
plot_search_time_vs_n(max_n, step)
```
